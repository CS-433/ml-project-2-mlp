```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Since the data is not provided in a csv format or similar, I'm recreating it based on the uploaded images.
# This is a simplified example and assumes that the categories are the same for both datasets and are sorted in the same order.

# Data from the images, manually entered.
data = {
    "Category": [
        "Business",
        "Society",
        "Arts",
        "Home",
        "Health",
        "Recreation",
        "Computers",
        "Shopping",
        "Reference",
        "Science",
        "Sports",
        "Kids and Teens",
        "News",
        "Games",
    ],
    "curlie-gpt3.5-10k": [
        36.7,
        25.7,
        19.5,
        13.5,
        11.8,
        9.9,
        8.8,
        7.8,
        6.4,
        6.0,
        4.9,
        3.2,
        2.7,
        2.5,
    ],
    "Curlie": [
        27.6,
        13.9,
        9.3,
        8.4,
        7.4,
        8.4,
        6.2,
        7.4,
        4.3,
        4.8,
        6.8,
        1.1,
        1.1,
        1.7,
    ],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to long format for seaborn
df_long = df.melt(id_vars="Category", var_name="Labeler", value_name="Percentage")

# Create the barplot with seaborn
plt.figure(figsize=(10, 8))
sns.barplot(x="Percentage", y="Category", hue="Labeler", data=df_long)

# Enhance the plot with labels and title
# plt.title(
#     "Comparison of Category Percentages in the curlie-gpt3.5-10k and original curlie dataset"
# )
plt.xlabel("Percentage")
plt.ylabel("Category")
plt.legend(title="Dataset")

# Display the plot
plt.savefig("../report/figures/class_distribution_comparison.pdf")

plt.show()
# save as pdf
```

```python
# Create the barplot with seaborn with improved aesthetics
plt.figure(figsize=(14, 10))
barplot = sns.barplot(
    x="Percentage",
    y="Category",
    hue="Labeler",
    data=df_long,
    palette="muted",
    edgecolor=".2",
)

# Increase font sizes for readability and set a style
sns.set_context("talk")  # This increases the font sizes
sns.set_style("whitegrid")  # This adds a grid and removes the top and right spines

# Add labels to the bars
for container in barplot.containers:
    barplot.bar_label(container, fmt="%.1f", label_type="edge", padding=3)

# Adjust the plot with labels and title
plt.title(
    "Comparison of Category Percentages in the Curlie-GPT-3.5-10k and Original Curlie Dataset"
)
plt.xlabel("Percentage")
plt.ylabel("Category")

# Move the legend outside of the plot to avoid blocking any bars
plt.legend(title="Labeler", bbox_to_anchor=(1.02, 1), loc="upper left", borderaxespad=0)

# Despine the plot for better aesthetics
sns.despine(trim=True)

# Adjust layout to make room for the legend and prevent clipping of tick-labels
plt.tight_layout()

# Display the plot
plt.show()
```
